# using matplot not seaborn

```python
import matplotlib.pyplot as plt 
import numpy as np
%matplotlib inline
```

```python
x = np.linspace(0, 15, 100)
```

```python
y1 = np.sin(x)
y2 = 2*np.sin(x+0.5)
y3 = 3*np.sin(x+1.0)
y4 = 4*np.sin(x+1.5)
```

```python
plt.figure(figsize=(10,6))
plt.plot(x,y1, x,y2, x,y3, x,y4)
plt.show()
```

# using seaborn

```python
import seaborn as sns

sns.set_style('darkgrid')
plt.figure(figsize=(10,6))
plt.plot(x,y1,x,y2,x,y3,x,y4)
plt.show()
```

# example1 with tips file

```python
tips = sns.load_dataset('tips')
tips.head(3)
#This data set organized by day of the week for lunch, dinner, smoking status, and meal amounts and tips.
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='day', y='total_bill', data = tips)
plt.show()
#This graph simply visualizes sales by day of the week.
```

```python
#There's another option called hue that allows you to differentiate your data using another data point
plt.figure(figsize=(8,6))
sns.boxplot(x='day',y='total_bill', hue='smoker', data = tips, palette='Set3')
plt.show()
```

If you check the graph above, you can see that smokers have a larger payment range

### Visualization regression analysis of payments and tips to employees

```python
sns.set_style('darkgrid')
sns.lmplot(x='total_bill', y='tip', data = tips, height = 4)
plt.show()
```

```python
#In conclusion, we can say that "total_bill and tip are positively correlated" 
#(the more total_bill, the more tip).
#Let's compare it to a smoker using HUE.
```

```python
sns.lmplot(x='total_bill', y='tip', hue='smoker', data=tips, palette='Set1', height=4)
plt.show()
```

# example2
Here's another data set on airline passengers by year and month

```python
flights = sns.load_dataset("flights")
flights.tail(3)
```

```python
# I'll utilize a pivot_table to break down the number of airplane passengers by year and month
```

```python
flights = flights.pivot(index = 'month', columns = 'year', values='passengers')
flights.head()
```

```python
# Let's see our data with hearmap func!
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(flights, annot=True, fmt='d')
plt.show()
```
